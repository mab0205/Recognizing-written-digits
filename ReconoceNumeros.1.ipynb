```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt

# libreria para matriz de confusion 
from sklearn import metrics 

%matplotlib inline
```

```python
#load_digits es una base d edatos de una linea que basicamente carga  todos los datos que necesitamos 

digits = load_digits()
```

```python
# la variable digitos genera datos. ne este caso son 1797 imagenes, de 8x8 y con un a dimension de 64
digits.data.shape
```

```python
#target son etiquetas (labels) que corresponden a cada imagen 
digits.target.shape 
```

```python
#tamanho del grafico que va a ser plotado 
plt.figure(figsize=(100,10))
#imprimir cada imagen
#digits.data quiere decir que vamos imprimir los primeros 5 digitos y target que numeros representan esos primeros 5 digitos 
for index, (image, label) in enumerate(zip(digits.data[0:15], digits.target[0:15])):
    #subplot se usa para plotar la imagen (numero de filas, numero de columnas, index)
 plt.subplot(1, 15, index + 1)
 # plt.cm.gray es para imprimir la imagen en una escala de grises 
 plt.imshow(np.reshape(image, (8,8)), cmap=plt.cm.gray)
 plt.title('Training: %i\n' % label, fontsize = 60)
```

Para modelo supervisado de ML, tenemos que dividir los datos en trainig and test sets. 
Training data para entender la relacion entre los accesorios, las imagenes y las etiquetas 
> Vamos a dividir los datos en 80% para aprendizaje y 20% para testing.  
> Uso de la libreria from sklearn.model_selection import train_test_split.

```python
#dividiendo los datos en 80 - 20... se usa la libreira train y test para dividir los datos 
x_train, x_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.20, random_state=0)
```

```python
print("tamanho de los datos de entrenamiento en X:",x_train.shape)
print("tamanho de los datos de entrenamiento en Y:",y_train.shape)
print("tamanho de los datos de test en X:",x_test.shape)
print("tamanho de los datos de test en Y:",y_test.shape)
print ("sumatoria de digitos de test y training: ", digits.data.shape)
```

### Patrón de modelado (Modeling Pattern (Digits Dataset))
En esta parte se importa el modelo que se quiere usar por ejemplo modelo de Regresion 


```python
#modelo de regresion logisticas
from sklearn.linear_model import LogisticRegression
# Todos los parametros no especificados son seteados a default
# Instanciando el modelo 
logisticRegr = LogisticRegression()
```

### **Entrenando el modelo sobre los datos y guardando la informacion a partir de los datos**
>Training the model on the data, storing the information learned from the data

```python
logisticRegr.fit(x_train, y_train)
```

```python
# Returns a NumPy Array
# Predict for One Observation (image)
logisticRegr.predict(x_test[70:85])
```

```python

for index, (image, label) in enumerate(zip(x_test[70:85], y_test[70:85])):
    #subplot se usa para plotar la imagen (numero de filas, numero de columnas, index)
 plt.subplot(12, 15, index + 1)
 # plt.cm.gray es para imprimir la imagen en una escala de grises 
 plt.imshow(np.reshape(image, (8,8)), cmap=plt.cm.gray)
 plt.title('Training: %i\n' % label, fontsize = 4)
```
